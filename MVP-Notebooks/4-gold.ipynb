```python
# Seleciona o catálogo e o schema gold
spark.sql("USE CATALOG mvp")
spark.sql("USE SCHEMA gold")
```

```python
# Carrega a tabela silver 'dados_postos_anp' do catálogo 'mvp'
df = spark.table("mvp.silver.dados_postos_anp_s")

# Salva o DataFrame como uma nova tabela 'dados_postos_anp_g' sobrescrevendo se já existir
df.write.mode("overwrite").saveAsTable("dados_postos_anp_g")

# Exibe uma amostra dos dados
display(df.limit(10))
```

```python
# Carrega as tabelas das três camadas
df_gold = spark.table("mvp.gold.dados_postos_anp_g")
df_silver = spark.table("mvp.silver.dados_postos_anp_s")
df_bronze = spark.table("mvp.bronze.dados_postos_anp")

# Compara o número de colunas e linhas entre ouro, prata e bronze
comparacao = {
    "gold": {
        "colunas": len(df_gold.columns),
        "linhas": df_gold.count()
    },
    "silver": {
        "colunas": len(df_silver.columns),
        "linhas": df_silver.count()
    },
    "bronze": {
        "colunas": len(df_bronze.columns),
        "linhas": df_bronze.count()
    }
}

display(comparacao)
```

```python
# Remove linhas duplicadas do DataFrame gold
df_gold_dedup = df_gold.dropDuplicates()

# Realiza nova contagem de linhas após remoção de duplicatas
nova_contagem = df_gold_dedup.count()

display({"linhas_gold_sem_duplicatas": nova_contagem})
```
